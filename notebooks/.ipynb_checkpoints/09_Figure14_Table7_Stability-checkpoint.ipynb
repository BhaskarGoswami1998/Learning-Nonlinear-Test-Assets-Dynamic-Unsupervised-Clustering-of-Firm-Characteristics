# Notebook 09 — Figure 14 & Table 7: Cluster Stability

**Data needed:** `data/firm_panel.csv` with `permno, year_month, cluster`

**Output:**
- Figure 14a: Within-window mean-reverting probability P(≤h)
- Figure 14b: Exact-month mean-reverting probability P(=h)
- Table 7: Freshness regression (new firms → next-month cluster return)

```
import os, sys, warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
warnings.filterwarnings("ignore")

REPO_ROOT = os.path.abspath(os.path.join(os.getcwd(), ".."))
sys.path.insert(0, REPO_ROOT)
from utils.data_utils import (load_cluster_panel, pivot_returns, rank_clusters,
                               load_firm_panel, nw_tstat, stars, save_figure,
                               save_table, GRID_DIR)

plt.rcParams.update({"font.family":"serif","font.size":10,
                      "axes.spines.top":False,"axes.spines.right":False})
```

```
# ── Load firm panel ───────────────────────────────────────────────
df_cl = load_cluster_panel(K=50, lam=1_000_000, data_dir=GRID_DIR)
cr    = pivot_returns(df_cl, dropna=True)
cr_r, rank_map = rank_clusters(cr)

try:
    firm = load_firm_panel()[["year_month","permno","cluster"]].dropna()
    firm["cluster_ranked"] = firm["cluster"].map(rank_map)
    firm = firm.sort_values(["permno","year_month"]).reset_index(drop=True)
    print(f"Firm panel loaded: {firm.shape}")
except FileNotFoundError:
    print("firm_panel.csv not found."); firm = None
```

```
# ── Mean-reverting probability ────────────────────────────────────
if firm is not None:
    H_MAX = 12
    
    # For each firm-month where firm leaves its cluster next month:
    # record baseline cluster and check if it returns within h months
    firm["next_cluster"] = firm.groupby("permno")["cluster_ranked"].shift(-1)
    leavers = firm[firm["cluster_ranked"] != firm["next_cluster"]].copy()
    leavers = leavers.dropna(subset=["next_cluster"])
    
    # Build forward cluster history per firm
    all_ym = sorted(firm["year_month"].unique())
    ym_idx = {ym: i for i, ym in enumerate(all_ym)}
    
    within_counts = np.zeros(H_MAX)
    exact_counts  = np.zeros(H_MAX)
    total = 0
    
    for _, row in leavers.iterrows():
        perm = row["permno"]
        t0   = row["year_month"]
        base = row["cluster_ranked"]
        t0_i = ym_idx.get(t0, -1)
        if t0_i < 0: continue
        
        future = firm[(firm["permno"]==perm) &
                      (firm["year_month"].apply(lambda x: ym_idx.get(x,-1)) > t0_i) &
                      (firm["year_month"].apply(lambda x: ym_idx.get(x,-1)) <= t0_i + H_MAX)]
        if len(future) == 0: continue
        
        total += 1
        returned = False
        for h in range(1, H_MAX + 1):
            h_i   = t0_i + h
            future_h = future[future["year_month"].apply(lambda x: ym_idx.get(x,-1)) == h_i]
            if len(future_h) > 0 and future_h["cluster_ranked"].iloc[0] == base:
                within_counts[h-1:] += 1
                if not returned:
                    exact_counts[h-1] += 1
                    returned = True
                break
            elif len(future_h) > 0:
                pass
        else:
            # Check cumulative
            for h in range(1, H_MAX + 1):
                h_i = t0_i + h
                fh  = future[future["year_month"].apply(lambda x: ym_idx.get(x,-1)) == h_i]
                if len(fh) > 0 and fh["cluster_ranked"].iloc[0] == base:
                    within_counts[h-1:] += 1
                    if not returned:
                        exact_counts[h-1] += 1
                        returned = True
    
    p_within = within_counts / max(total, 1)
    p_exact  = exact_counts  / max(total, 1)
    print(f"Processed {total:,} leaver events")
    print(f"P(≤12): {p_within[-1]:.3f}  (paper: ~0.85)")
    print(f"P(=1):  {p_exact[0]:.3f}   (paper: ~0.58)")
```

```
# ── Figure 14 ─────────────────────────────────────────────────────
if firm is not None:
    hs = np.arange(1, H_MAX + 1)
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(11, 4.5))
    
    ax1.plot(hs, p_within, "o-", color="black", linewidth=2, markersize=7, markerfacecolor="white")
    ax1.set_xlabel("Horizon h (months)"); ax1.set_ylabel("Probability")
    ax1.set_title("(a) Mean-Reverting Probability P(≤h)", fontweight="bold")
    ax1.set_xticks(hs); ax1.set_xlim(0.5, H_MAX + 0.5); ax1.set_ylim(0, 1)
    ax1.grid(True, alpha=0.3)
    
    ax2.plot(hs, p_exact, "o-", color="black", linewidth=2, markersize=7, markerfacecolor="white")
    ax2.set_xlabel("Horizon h (months)"); ax2.set_ylabel("Probability")
    ax2.set_title("(b) Exact-Month Mean-Reverting Probability P(=h)", fontweight="bold")
    ax2.set_xticks(hs); ax2.set_xlim(0.5, H_MAX + 0.5)
    ax2.grid(True, alpha=0.3)
    
    fig.suptitle("Figure 14: Mean-Reverting Probabilities", fontsize=12, fontweight="bold")
    plt.tight_layout()
    save_figure(fig, "Figure14_Mean_Reverting_Probability")
    plt.show()
```

```
# ── Table 7: Freshness regression ────────────────────────────────
if firm is not None:
    # Freshness = fraction of firms in cluster that are "new" (not there last month)
    firm_sorted = firm.sort_values(["permno","year_month"])
    firm_sorted["prev_cluster"] = firm_sorted.groupby("permno")["cluster_ranked"].shift(1)
    firm_sorted["is_new"] = (firm_sorted["cluster_ranked"] != firm_sorted["prev_cluster"]) |                              firm_sorted["prev_cluster"].isna()
    
    freshness = firm_sorted.groupby(["year_month","cluster_ranked"])["is_new"].mean().reset_index()
    freshness.columns = ["year_month","cluster_ranked","freshness"]
    
    # Merge with next-month cluster return
    ret_long = df_cl[["year_month","cluster","ret"]].copy()
    ret_long["cluster_ranked"] = ret_long["cluster"].map(rank_map)
    ret_long["date"] = pd.to_datetime(ret_long["year_month"], format="%Y-%m")
    ret_long["year_month_next"] = (ret_long["date"] + pd.DateOffset(months=1)).dt.strftime("%Y-%m")
    
    panel = freshness.merge(
        ret_long[["year_month","cluster_ranked","ret"]].rename(
            columns={"year_month":"year_month_next","ret":"ret_next"}),
        left_on=["year_month","cluster_ranked"],
        right_on=["year_month_next","cluster_ranked"], how="inner"
    ).dropna(subset=["freshness","ret_next"])
    
    panel["time_id"]    = panel["year_month"].astype(str)
    panel["cluster_id"] = panel["cluster_ranked"].astype(str)
    
    # Spec (1): TWFE no controls
    def twfe_reg(y, x, time_id, cluster_id):
        df = pd.DataFrame({"y":y,"x":x,"t":time_id,"c":cluster_id}).dropna()
        df["y"] -= df.groupby("t")["y"].transform("mean")
        df["y"] -= df.groupby("c")["y"].transform("mean")
        df["x"] -= df.groupby("t")["x"].transform("mean")
        df["x"] -= df.groupby("c")["x"].transform("mean")
        X = sm.add_constant(df["x"].values)
        res = sm.OLS(df["y"].values, X).fit(cov_type="cluster",
              cov_kwds={"groups": df["c"].values})
        return res.params[1], res.bse[1], res.tvalues[1], len(df), res.rsquared
    
    coef1, se1, t1, n1, r2_1 = twfe_reg(panel["ret_next"], panel["freshness"],
                                          panel["time_id"], panel["cluster_id"])
    
    table7 = pd.DataFrame({
        "(1)": [f"{coef1:.4f}{stars(t1)}", f"({t1:.2f})", "No", "Yes", "Yes",
                f"{r2_1:.4f}", f"{n1:,}"],
    }, index=["Freshness","t-stat","Controls","Time FE","Cluster FE","R²","N"])
    
    print("\n=== Table 7 ===")
    print(table7.to_string())
    print(f"\nPaper: coef=0.0167***, t=6.15  |  Ours: coef={coef1:.4f}, t={t1:.2f}")
    save_table(table7, "Table7_Freshness_Regression")
    print("Table 7 saved.")
```

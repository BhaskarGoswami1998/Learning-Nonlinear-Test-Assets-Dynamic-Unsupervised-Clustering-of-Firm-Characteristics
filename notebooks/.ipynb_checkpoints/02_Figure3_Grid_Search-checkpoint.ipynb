# Notebook 02 — Figure 3: H-L Return and SDF Sharpe Ratio Across K and λ

**Data needed:** All `data/grid/cluster_month_panel_K_{K}_lambda_{λ}.csv` files  
**Output:** `output/figures/Figure3_HL_Heatmap.{pdf,png}`, `Figure3_SDF_SR.{pdf,png}`

For each (K, λ) combination this notebook:
1. Loads the cluster panel
2. Pivots to return matrix, ranks clusters
3. Computes annualised H-L return (long L_max, short L_min)
4. Estimates the rolling HJ-SDF and its Sharpe ratio
5. Plots results as line charts (K on x-axis, λ as colored lines)

```
import os, sys, warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
warnings.filterwarnings("ignore")

REPO_ROOT = os.path.abspath(os.path.join(os.getcwd(), ".."))
sys.path.insert(0, REPO_ROOT)
from utils.data_utils import (load_cluster_panel, pivot_returns, rank_clusters,
                               ann_sharpe, save_figure, GRID_DIR,
                               K_GRID, LAMBDA_GRID, lambda_to_filestr)
from utils.portfolio_utils import estimate_hj_sdf

plt.rcParams.update({"font.family":"serif","font.size":10,
                      "axes.spines.top":False,"axes.spines.right":False})

SDF_WINDOW = 72   # rolling window for SDF estimation (months)
RIDGE      = 1e-3
```

```
# ── Grid scan ─────────────────────────────────────────────────────
# Results stored as DataFrames indexed by K (rows) × λ string (cols)
lam_labels = [f"λ={lambda_to_filestr(l)}" for l in LAMBDA_GRID]

hl_mat  = pd.DataFrame(index=K_GRID, columns=lam_labels, dtype=float)
sr_mat  = pd.DataFrame(index=K_GRID, columns=lam_labels, dtype=float)

print("Scanning grid... (this may take several minutes)")
for K in K_GRID:
    for lam, lab in zip(LAMBDA_GRID, lam_labels):
        try:
            df  = load_cluster_panel(K=K, lam=lam, data_dir=GRID_DIR)
            cr  = pivot_returns(df, dropna=True)
            cr_r, _ = rank_clusters(cr)

            # ── H-L annualised return ────────────────────────────────
            hl = cr_r.iloc[:, -1] - cr_r.iloc[:, 0]
            hl_ann = (np.prod(1 + hl.dropna()) ** (12 / len(hl.dropna())) - 1) * 100
            hl_mat.loc[K, lab] = hl_ann

            # ── Rolling SDF Sharpe ───────────────────────────────────
            T = len(cr_r)
            sdf_rets = []
            for t in range(SDF_WINDOW, T - 1):
                train = cr_r.iloc[t - SDF_WINDOW: t]
                try:
                    b, _ = estimate_hj_sdf(train, ridge=RIDGE)
                    r_next = cr_r.iloc[t + 1].values
                    sdf_rets.append(1.0 - b @ r_next)
                except Exception:
                    pass
            if sdf_rets:
                s = pd.Series(sdf_rets)
                sr_mat.loc[K, lab] = ann_sharpe(s)
        except FileNotFoundError:
            pass   # file not available yet

print("Grid scan complete.")
print(f"H-L matrix: {hl_mat.notna().sum().sum()} / {hl_mat.size} cells filled")
print(f"SDF-SR matrix: {sr_mat.notna().sum().sum()} / {sr_mat.size} cells filled")
```

```
# ── Save grid results for later use ───────────────────────────────
import os
os.makedirs(os.path.join(REPO_ROOT,"data"), exist_ok=True)
hl_mat.to_csv(os.path.join(REPO_ROOT,"data","grid_HL_results.csv"))
sr_mat.to_csv(os.path.join(REPO_ROOT,"data","grid_SDF_SR_results.csv"))
print("Grid results saved to data/grid_HL_results.csv and data/grid_SDF_SR_results.csv")
```

```
# ── Figure 3a: H-L line chart ─────────────────────────────────────
# Match paper style: K on x-axis, one colored line per λ
# Only plot λ values that have data
cmap   = cm.get_cmap("tab20", len(LAMBDA_GRID))
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

ax = axes[0]
for j, (lam, lab) in enumerate(zip(LAMBDA_GRID, lam_labels)):
    row = hl_mat[lab].dropna()
    if len(row) > 0:
        ax.plot(row.index, row.values, color=cmap(j),
                linewidth=1.5, marker="o", markersize=3,
                label=f"λ={lambda_to_filestr(lam)}")

ax.axvline(50, color="black", linestyle="--", linewidth=0.8, alpha=0.6)
ax.set_xlabel("Number of clusters K")
ax.set_ylabel("Annualised H-L return (%)")
ax.set_title("(a) High-minus-Low Portfolio Return", fontweight="bold")
ax.legend(title="λ", ncol=2, fontsize=7, loc="upper left",
          bbox_to_anchor=(0, 1), framealpha=0.9)

# Figure 3b: SDF SR line chart
ax = axes[1]
for j, (lam, lab) in enumerate(zip(LAMBDA_GRID, lam_labels)):
    row = sr_mat[lab].dropna()
    if len(row) > 0:
        ax.plot(row.index, row.values, color=cmap(j),
                linewidth=1.5, marker="o", markersize=3,
                label=f"λ={lambda_to_filestr(lam)}")

ax.axvline(50, color="black", linestyle="--", linewidth=0.8, alpha=0.6)
ax.set_xlabel("Number of clusters K")
ax.set_ylabel("Annualised SDF Sharpe ratio")
ax.set_title("(b) SDF Sharpe Ratio", fontweight="bold")
ax.legend(title="λ", ncol=2, fontsize=7, loc="upper right",
          bbox_to_anchor=(1, 1), framealpha=0.9)

fig.suptitle("Figure 3: Out-of-Sample Performance Across Hyperparameter Values",
             fontsize=12, fontweight="bold")
plt.tight_layout()
save_figure(fig, "Figure3_HL_and_SDF_SR")
plt.show()
print("Figure 3 saved.")
```

# Notebook 00 — Data Setup & Directory Guide

Run this notebook **first**. It checks that all required data files are present
and explains exactly where each file should go.

## Required directory structure

```
data/
├── grid/                          ← ALL cluster-month panels go here
│   ├── cluster_month_panel_K_2_lambda_1e-09.csv
│   ├── cluster_month_panel_K_2_lambda_1e-08.csv
│   │   ... (one file per K × λ combination)
│   └── cluster_month_panel_K_100_lambda_1000000000.csv
│
├── factor_returns.csv             ← Monthly factor returns (see below)
├── macro_predictors.csv           ← Goyal-Welch macro variables (see below)
├── mebm_25.csv                    ← Fama-French 25 ME-BM portfolios
├── benchmark_sorts.csv            ← Uni-Sort 150, Bi-Sort 285, Ind49
├── centroid_chars.csv             ← Cluster centroid characteristics (your output)
└── firm_panel.csv                 ← Firm-level panel (your output)
```

## `factor_returns.csv` format
Columns: `date, MKT, SMB, HML, RMW, CMA, MOM, LIQ, PEAD, FIN, IA, ROE_q, EG, MGMT, PERF, RF`  
Sources: Ken French, Lubos Pástor, Lu Zhang (q-library), Kent Daniel, Robert Stambaugh

## `macro_predictors.csv` format
Columns: `date, dp, dy, ep, tbl, tms, dfy, infl, svar, lev, ni`  
Source: Goyal-Welch PredictorData2024.xlsx (Amit Goyal's website)

## `centroid_chars.csv` format
Columns: `year_month, cluster, A2ME, AC, AT, ..., VAR`  (47 columns total)  
This is your own clustering output: the value-weighted mean of each of the 45 firm
characteristics within each cluster, for each month.

## `firm_panel.csv` format
Columns: `year_month, permno, cluster, ret, me, SIC, A2ME, AC, ..., VAR`  
This is your full firm-level panel with cluster assignments, returns, market cap,
SIC code, and all 45 characteristics.

````
import os, sys
import pandas as pd
import numpy as np

REPO_ROOT = os.path.abspath(os.path.join(os.getcwd(), ".."))
sys.path.insert(0, REPO_ROOT)
from utils.data_utils import (GRID_DIR, DATA_DIR, K_GRID, LAMBDA_GRID,
                               lambda_to_filestr, load_cluster_panel)

print("=" * 60)
print("DATA SETUP CHECKER")
print("=" * 60)
````

````
# ── 1. Check grid files ──────────────────────────────────────────
print("\n1. GRID FILES (cluster-month panels)")
print(f"   Expected location: {GRID_DIR}")
os.makedirs(GRID_DIR, exist_ok=True)

total_expected = len(K_GRID) * len(LAMBDA_GRID)
found = []
missing = []

for K in K_GRID:
    for lam in LAMBDA_GRID:
        fname = f"cluster_month_panel_K_{K}_lambda_{lambda_to_filestr(lam)}.csv"
        path  = os.path.join(GRID_DIR, fname)
        if os.path.exists(path):
            found.append((K, lam))
        else:
            missing.append(fname)

print(f"   Found  : {len(found)} / {total_expected}")
print(f"   Missing: {len(missing)}")
if missing[:5]:
    print("   First 5 missing:")
    for m in missing[:5]: print(f"     {m}")

# Check K=50, λ=1e6 specifically (baseline)
baseline = os.path.join(GRID_DIR, "cluster_month_panel_K_50_lambda_1000000.csv")
if os.path.exists(baseline):
    df = pd.read_csv(baseline)
    print(f"\n   ✓ Baseline (K=50, λ=1e6): {df.shape[0]:,} rows, "
          f"{df['cluster'].nunique()} clusters, "
          f"{df['year_month'].nunique()} months")
else:
    print("\n   ✗ BASELINE FILE MISSING — copy it to data/grid/")
````

````
# ── 2. Check auxiliary files ─────────────────────────────────────
print("\n2. AUXILIARY FILES")
aux_files = {
    "factor_returns.csv":   "Factor returns (FF5, Q5, DHS, etc.)",
    "macro_predictors.csv": "Goyal-Welch macro predictors",
    "mebm_25.csv":          "Fama-French 25 ME-BM portfolios",
    "benchmark_sorts.csv":  "Uni-Sort, Bi-Sort, Ind49 portfolios",
    "centroid_chars.csv":   "Cluster centroid characteristics",
    "firm_panel.csv":       "Firm-level panel with SIC & chars",
}
for fname, desc in aux_files.items():
    path   = os.path.join(DATA_DIR, fname)
    status = "✓" if os.path.exists(path) else "✗ MISSING"
    print(f"   {status}  {fname:30s} — {desc}")
````

````
# ── 3. Summary: which notebooks can run ────────────────────────────
print("\n3. NOTEBOOK AVAILABILITY")
print("-" * 55)
has_grid     = len(found) > 0
has_baseline = os.path.exists(os.path.join(GRID_DIR,"cluster_month_panel_K_50_lambda_1000000.csv"))
has_factors  = os.path.exists(os.path.join(DATA_DIR, "factor_returns.csv"))
has_macro    = os.path.exists(os.path.join(DATA_DIR, "macro_predictors.csv"))
has_mebm     = os.path.exists(os.path.join(DATA_DIR, "mebm_25.csv"))
has_bench    = os.path.exists(os.path.join(DATA_DIR, "benchmark_sorts.csv"))
has_centroid = os.path.exists(os.path.join(DATA_DIR, "centroid_chars.csv"))
has_firm     = os.path.exists(os.path.join(DATA_DIR, "firm_panel.csv"))

notebooks = [
    ("01_Table1", "Table 1 — Cluster Returns",           has_baseline),
    ("02_Figure3","Figure 3 — H-L & SDF SR Heatmaps",    has_grid),
    ("03_Figures4_5_6","Figures 4,5,6 — SDF Diagnostics",has_baseline),
    ("04_Tables2_3","Tables 2&3 — Spanning Tests & H-L",  has_baseline and has_factors),
    ("05_Table4","Table 4 — HML Double-Sort",             has_firm),
    ("06_Table5","Table 5 — Panel OLS TWFE",              has_centroid),
    ("07_Table6","Table 6 — Differential Entropy",        has_centroid),
    ("08_Figures8_9","Figures 8&9 — ML Feature Importance",has_centroid),
    ("09_Figures7_8","Figures 7&8 — Industry Heatmaps",   has_firm),
    ("10_Figures9_10","Figures 9&10 — Char Profiles",     has_centroid),
    ("11_Figures15_16","Figures 15&16 — MVE Frontiers",   has_baseline and has_factors and has_mebm),
    ("12_Table8","Table 8 — Macro Regime Switches",        has_baseline and has_factors and has_macro),
    ("13_Figure14","Figure 14 — Mean-Revert Probability",  has_firm),
]
for nb_id, desc, ready in notebooks:
    icon = "✓ READY" if ready else "✗ needs more data"
    print(f"  {icon:12s}  {desc}")
print()
print("Copy your CSV files to the locations shown above, then re-run this notebook.")
````

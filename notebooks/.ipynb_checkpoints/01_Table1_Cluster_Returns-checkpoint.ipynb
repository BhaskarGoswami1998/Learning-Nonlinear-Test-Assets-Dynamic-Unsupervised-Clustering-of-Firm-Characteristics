# Notebook 01 — Table 1: Monthly Out-of-Sample Value-Weighted Cluster Returns

**Data needed:** `data/grid/cluster_month_panel_K_50_lambda_1000000.csv`  
**Output:** `output/tables/Table1_Cluster_Returns.{csv,tex}`, bar-chart figure

Replicates Table 1 exactly: Mean Return, Std Dev, Sharpe, Max Drawdown, Calmar, VaR(95%)
for all 50 value-weighted cluster portfolios, with Newey-West t-statistics (3 lags).

```
import os, sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm

REPO_ROOT = os.path.abspath(os.path.join(os.getcwd(), ".."))
sys.path.insert(0, REPO_ROOT)
from utils.data_utils import (load_cluster_panel, pivot_returns, rank_clusters,
                               nw_tstat, stars, ann_sharpe, max_dd, calmar,
                               var95, save_table, save_figure, GRID_DIR)

plt.rcParams.update({"font.family":"serif","font.size":10,
                      "axes.spines.top":False,"axes.spines.right":False})
NW_LAGS = 3
```

```
# ── Load baseline panel ───────────────────────────────────────────
df  = load_cluster_panel(K=50, lam=1_000_000, data_dir=GRID_DIR)
cr  = pivot_returns(df, col="ret", dropna=True)   # (T x 50) matrix
cr_ranked, rank_map = rank_clusters(cr)

print(f"Cluster return matrix: {cr_ranked.shape[0]} months × {cr_ranked.shape[1]} clusters")
print(f"Sample: {cr_ranked.index[0].strftime('%Y-%m')} → {cr_ranked.index[-1].strftime('%Y-%m')}")
```

```
# ── Compute Table 1 statistics ────────────────────────────────────
rows = []
for label in cr_ranked.columns:          # L01 … L50
    r = cr_ranked[label].dropna()
    mean_r, se, t = nw_tstat(r, lags=NW_LAGS)
    rows.append({
        "Cluster":       label,
        "Mean Ret":      round(mean_r, 4),
        "Sig":           stars(t),
        "t-stat":        round(t, 4),
        "Std Dev":       round(r.std(), 4),
        "Sharpe":        round(ann_sharpe(r), 4),
        "Max Drawdown":  round(abs(max_dd(r)), 4),
        "Calmar":        round(calmar(r), 4),
        "VaR (95%)":     round(var95(r), 4),
    })

table1 = pd.DataFrame(rows).set_index("Cluster")
print(table1[["Mean Ret","Sig","Std Dev","Sharpe","Max Drawdown","Calmar","VaR (95%)"]].to_string())
```

```
# ── Verify key paper numbers ──────────────────────────────────────
print("=== Paper verification ===")
print(f"L01 Mean Ret:  {table1.loc['L01','Mean Ret']:.4f}  (paper: -0.0185)")
print(f"L50 Mean Ret:  {table1.loc['L50','Mean Ret']:.4f}  (paper:  0.0158)")
print(f"L01 Sharpe:    {table1.loc['L01','Sharpe']:.4f}  (paper: -0.5997)")
print(f"L48 Sharpe:    {table1.loc['L48','Sharpe']:.4f}  (paper:  1.0139)")
print(f"L26 Max DD:    {table1.loc['L26','Max Drawdown']:.4f}  (paper:  0.3298)")
```

```
# ── Save table ────────────────────────────────────────────────────
out_cols = ["Mean Ret","Sig","Std Dev","Sharpe","Max Drawdown","Calmar","VaR (95%)"]
save_table(table1[out_cols], "Table1_Cluster_Returns")

# ── Bar chart ─────────────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(13, 3.8))
labels = table1.index.tolist()
vals   = table1["Mean Ret"].values * 100
colors = cm.RdYlGn(np.linspace(0.05, 0.95, len(vals)))

bars = ax.bar(labels, vals, color=colors, edgecolor="white", linewidth=0.3)
ax.axhline(0, color="black", linewidth=0.8, linestyle="--")
ax.set_xlabel("Cluster (ranked by mean return)", fontsize=10)
ax.set_ylabel("Mean excess return (% / month)", fontsize=10)
ax.set_title("Table 1: Monthly Out-of-Sample Value-Weighted Cluster Returns",
             fontsize=11, fontweight="bold")

# Label every 5th cluster
for i, (bar, lbl) in enumerate(zip(bars, labels)):
    if i % 5 == 0:
        ax.text(bar.get_x() + bar.get_width()/2, -0.003 if vals[i] >= 0 else vals[i] - 0.015,
                lbl, ha="center", va="top", fontsize=7, rotation=90)

# Suppress x-tick labels (too crowded)
ax.set_xticks([])
plt.tight_layout()
save_figure(fig, "Table1_Cluster_Returns_Bar")
plt.show()
print("Done.")
```

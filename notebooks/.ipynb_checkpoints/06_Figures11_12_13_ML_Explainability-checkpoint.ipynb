# Notebook 06 — Figures 11, 12, 13: ML Feature Importance & Decision Trees

**Data needed:** `data/centroid_chars.csv`

**Output:**
- Figure 11a: Entropy-based feature importance heatmap (DT, RF, XGB, LGBM)
- Figure 11b: Value-based feature importance heatmap
- Figure 12: Decision tree for post-2002 subsample
- Figure 13: Decision tree for pre-2002 subsample

```
import os, sys, warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
warnings.filterwarnings("ignore")

try:
    from xgboost import XGBClassifier; HAS_XGB = True
except ImportError:
    HAS_XGB = False; print("xgboost not installed — skipping XGB")
try:
    from lightgbm import LGBMClassifier; HAS_LGBM = True
except ImportError:
    HAS_LGBM = False; print("lightgbm not installed — skipping LGBM")

REPO_ROOT = os.path.abspath(os.path.join(os.getcwd(), ".."))
sys.path.insert(0, REPO_ROOT)
from utils.data_utils import (load_cluster_panel, pivot_returns, rank_clusters,
                               load_centroid_chars, save_figure, GRID_DIR,
                               DATA_DIR, CHARS_45)

plt.rcParams.update({"font.family":"serif","font.size":9})
np.random.seed(42)
```

```
# ── Load centroid chars + rank clusters ──────────────────────────
df = load_cluster_panel(K=50, lam=1_000_000, data_dir=GRID_DIR)
cr = pivot_returns(df, dropna=True)
cr_ranked, rank_map = rank_clusters(cr)

try:
    cent = load_centroid_chars()
    cent["cluster_ranked"] = cent["cluster"].map(rank_map)
    char_cols = [c for c in CHARS_45 if c in cent.columns]
    le = LabelEncoder()
    cent["label"] = le.fit_transform(cent["cluster_ranked"].fillna("L00"))
    print(f"Centroid chars: {cent.shape}, {len(char_cols)} features")
except FileNotFoundError:
    print("centroid_chars.csv not found — cannot produce Figures 11-13.")
    cent = None; char_cols = []
```

```
# ── Build entropy-based and value-based feature matrices ─────────
def gaussian_entropy(sigma2):
    return 0.5 * np.log(2 * np.pi * np.e * np.maximum(sigma2, 1e-20))

if cent is not None:
    entropy_rows, value_rows = [], []
    for (dt, cl), grp in cent.groupby(["year_month","cluster_ranked"]):
        erow = {"year_month":dt,"cluster":cl,"label":grp["label"].iloc[0]}
        vrow = {"year_month":dt,"cluster":cl,"label":grp["label"].iloc[0]}
        for char in char_cols:
            vals = grp[char].dropna()
            erow[char] = gaussian_entropy(vals.var(ddof=1)) if len(vals)>=3 else np.nan
            vrow[char] = vals.mean() if len(vals)>=1 else np.nan
        entropy_rows.append(erow)
        value_rows.append(vrow)
    
    ent_df = pd.DataFrame(entropy_rows).dropna(subset=char_cols)
    val_df = pd.DataFrame(value_rows).dropna(subset=char_cols)
    print(f"Entropy features: {ent_df.shape}, Value features: {val_df.shape}")
```

```
# ── Train classifiers, get feature importance ranks ──────────────
def get_ranks(df, feats, target):
    X = df[feats].values; y = df[target].values
    classifiers = {"DT": DecisionTreeClassifier(max_depth=7,random_state=42),
                   "RF": RandomForestClassifier(n_estimators=300,max_depth=8,
                                                random_state=42,n_jobs=-1)}
    if HAS_XGB:
        classifiers["XGB"] = XGBClassifier(n_estimators=300,max_depth=5,
                                            random_state=42,verbosity=0,
                                            eval_metric="mlogloss")
    if HAS_LGBM:
        classifiers["LGBM"] = LGBMClassifier(n_estimators=300,max_depth=5,
                                              random_state=42,verbose=-1)
    rank_df = {}
    for name, clf in classifiers.items():
        clf.fit(X, y)
        imp  = pd.Series(clf.feature_importances_, index=feats)
        rank_df[name] = imp.rank(ascending=False).astype(int)
    return pd.DataFrame(rank_df)

if cent is not None:
    print("Training classifiers on entropy features...")
    ranks_ent = get_ranks(ent_df, char_cols, "label")
    print("Training classifiers on value features...")
    ranks_val = get_ranks(val_df, char_cols, "label")
    print("Done.")
```

```
# ── Figure 11: Feature importance heatmaps ───────────────────────
def plot_importance_heatmap(rank_df, title, fname):
    mean_rank = rank_df.mean(axis=1).sort_values()
    sdf = rank_df.loc[mean_rank.index]
    fig, ax = plt.subplots(figsize=(5, 11))
    sns.heatmap(sdf, ax=ax, cmap="YlOrRd_r", annot=True, fmt="d",
                annot_kws={"size":6}, linewidths=0.3, linecolor="white",
                cbar_kws={"label":"Rank (1=most important)"})
    ax.set_title(title, fontsize=10, fontweight="bold", pad=8)
    ax.set_xlabel("Classifier"); ax.set_ylabel("")
    ax.tick_params(axis="y", labelsize=7)
    plt.tight_layout()
    save_figure(fig, fname)
    return fig

if cent is not None:
    fig11a = plot_importance_heatmap(
        ranks_ent,
        "(a) Entropy-Based Feature Importance",
        "Figure11a_ML_Importance_Entropy"
    )
    fig11b = plot_importance_heatmap(
        ranks_val,
        "(b) Value-Based Feature Importance",
        "Figure11b_ML_Importance_Values"
    )
    plt.show()
    print("Figures 11a and 11b saved.")
```

```
# ── Figures 12 & 13: Decision trees pre/post 2002 ────────────────
if cent is not None:
    dt_clf = DecisionTreeClassifier(max_depth=5, random_state=42)
    
    for period, label in [("post_2002","≥2002"), ("pre_2002","<2002")]:
        if period == "post_2002":
            mask = pd.to_datetime(ent_df["year_month"]).dt.year >= 2002
        else:
            mask = pd.to_datetime(ent_df["year_month"]).dt.year < 2002
        
        sub = ent_df[mask]
        if len(sub) < 50: continue
        X = sub[char_cols].values; y = sub["label"].values
        dt_clf.fit(X, y)
        
        fig_dt, ax_dt = plt.subplots(figsize=(28, 10))
        plot_tree(dt_clf, feature_names=char_cols, ax=ax_dt,
                  class_names=[f"L{i+1:02d}" for i in range(len(le.classes_))],
                  filled=True, rounded=True, fontsize=7, max_depth=4)
        fig_num = "12" if period=="post_2002" else "13"
        title   = f"Figure {fig_num}: Decision Tree (Entropy) — {'Post' if period=='post_2002' else 'Pre'}-2002"
        ax_dt.set_title(title, fontsize=12, fontweight="bold")
        plt.tight_layout()
        save_figure(fig_dt, f"Figure{fig_num}_DecisionTree_{period}")
        plt.show()
        print(f"Figure {fig_num} saved.")
```

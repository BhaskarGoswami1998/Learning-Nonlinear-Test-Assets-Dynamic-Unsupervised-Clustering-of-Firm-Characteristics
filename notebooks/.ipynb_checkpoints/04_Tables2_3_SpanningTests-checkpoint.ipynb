# Notebook 04 — Tables 2 & 3: Factor Spanning Tests and H-L Comparison

**Data needed:**
- `data/grid/cluster_month_panel_K_50_lambda_1000000.csv`
- `data/factor_returns.csv`

**Output:**
- `Table2_Alpha_Diagnostics.{csv,tex}`
- `Table3_HL_Comparison.{csv,tex}`

**Table 2** replicates the four diagnostics (A|α|, A|α/r|, R², Sh²(α)) for 11 models.  
**Table 3** replicates H-L returns for Our Model, Vanilla, Inertia, and benchmark sorts.

```
import os, sys, warnings
import numpy as np
import pandas as pd
import statsmodels.api as sm
warnings.filterwarnings("ignore")

REPO_ROOT = os.path.abspath(os.path.join(os.getcwd(), ".."))
sys.path.insert(0, REPO_ROOT)
from utils.data_utils import (load_cluster_panel, pivot_returns, rank_clusters,
                               nw_tstat, stars, ann_sharpe, save_table, GRID_DIR, DATA_DIR)
from utils.portfolio_utils import estimate_hj_sdf

NW_LAGS = 3

FACTOR_MODELS = {
    "CAPM":  ["MKT"],
    "FF3":   ["MKT","SMB","HML"],
    "FFC":   ["MKT","SMB","HML","MOM"],
    "FF5":   ["MKT","SMB","HML","RMW","CMA"],
    "FFPS":  ["MKT","SMB","HML","LIQ"],
    "DHS":   ["MKT","PEAD","FIN"],
    "Q":     ["MKT","SMB","IA","ROE_q","EG"],
    "SYM":   ["MKT","SMB","MGMT","PERF"],
    "PCA":   ["PC1","PC2","PC3","PC4","PC5"],
    "IPCA":  ["IPC1","IPC2","IPC3","IPC4","IPC5"],
    "RPPCA": ["RPC1","RPC2","RPC3","RPC4","RPC5"],
}
```

```
# ── Load data ─────────────────────────────────────────────────────
df = load_cluster_panel(K=50, lam=1_000_000, data_dir=GRID_DIR)
cr = pivot_returns(df, dropna=True)
cr_ranked, _ = rank_clusters(cr)

try:
    factors = pd.read_csv(os.path.join(DATA_DIR, "factor_returns.csv"),
                          parse_dates=["date"]).set_index("date")
    print("Factor data loaded:", list(factors.columns))
except FileNotFoundError:
    print("factor_returns.csv not found — Table 2 will show NaN for all models.")
    print("Create data/factor_returns.csv with columns: date, MKT, SMB, HML, ...")
    factors = pd.DataFrame(index=cr_ranked.index)

common = cr_ranked.index.intersection(factors.index)
cr_c   = cr_ranked.loc[common]
fd_c   = factors.loc[common]
print(f"Common period: {common[0].date()} → {common[-1].date()} ({len(common)} months)")
```

```
# ── Helper: time-series regression with NW SEs ────────────────────
def ts_reg(y, X_df, lags=3):
    """Returns alpha, alpha_se, t_stat, R2, residuals."""
    idx = y.index.intersection(X_df.index)
    y_  = y.loc[idx].values
    X_  = sm.add_constant(X_df.loc[idx].values)
    if len(y_) < 24:
        return np.nan, np.nan, np.nan, np.nan, np.full(len(y_), np.nan)
    res = sm.OLS(y_, X_).fit(cov_type="HAC", cov_kwds={"maxlags": lags})
    return res.params[0], res.bse[0], res.tvalues[0], res.rsquared, res.resid
```

```
# ── Table 2: Alpha diagnostics across factor models ──────────────
table2_rows = {}
for model, cols in FACTOR_MODELS.items():
    avail = [c for c in cols if c in fd_c.columns]
    if not avail:
        print(f"  Skipping {model}: no columns found in factor data")
        continue
    if len(avail) < len(cols):
        print(f"  {model}: using {avail} (missing {set(cols)-set(avail)})")
    
    alphas, r2s, mean_rets = [], [], []
    for col in cr_c.columns:
        a, _, _, r2, _ = ts_reg(cr_c[col], fd_c[avail], lags=NW_LAGS)
        alphas.append(a); r2s.append(r2); mean_rets.append(cr_c[col].mean())
    
    alphas    = np.array(alphas)
    mean_rets = np.array(mean_rets)
    finite    = np.isfinite(alphas)
    sh2 = (np.mean(alphas[finite])/np.std(alphas[finite]))**2 if np.std(alphas[finite])>0 else np.nan
    
    table2_rows[model] = {
        "A|α_i|":     np.mean(np.abs(alphas[finite])),
        "A|α_i/r_i|": np.mean(np.abs(alphas[finite]/(mean_rets[finite]+1e-10))),
        "R²":          np.mean(np.array(r2s)[finite]),
        "Sh²(α)":     sh2,
    }

table2 = pd.DataFrame(table2_rows).T.round(4)
print("\n=== Table 2 ===")
print(table2.to_string())
save_table(table2, "Table2_Alpha_Diagnostics")
```

```
# ── Table 3: H-L portfolio comparison ────────────────────────────
# Our model H-L
hl_ours = cr_ranked.iloc[:, -1] - cr_ranked.iloc[:, 0]
high_   = cr_ranked.iloc[:, -1]
low_    = cr_ranked.iloc[:, 0]

def hl_row(name, low_ret, high_ret, hl_ret, nw_lags=3):
    m_l, se_l, t_l = nw_tstat(low_ret,  lags=nw_lags)
    m_h, se_h, t_h = nw_tstat(high_ret, lags=nw_lags)
    m_hl,se_hl,t_hl= nw_tstat(hl_ret,   lags=nw_lags)
    sr_hl = m_hl / hl_ret.std() if hl_ret.std()>0 else np.nan
    return {
        "": name,
        "Low":  f"{m_l:.4f}{stars(t_l)}",
        "t_Low":  f"({t_l:.4f})",
        "High": f"{m_h:.4f}{stars(t_h)}",
        "t_High": f"({t_h:.4f})",
        "H-L":  f"{m_hl:.4f}{stars(t_hl)}",
        "t_HL":   f"({t_hl:.4f})",
        "S(H-L)": f"{sr_hl:.4f}",
    }

rows3 = [hl_row("Our Model", low_, high_, hl_ours)]

# Vanilla and Inertia panels (other K=50 lambda files)
for lam_v, label_v in [(0.001,"Vanilla"),(100000,"Inertia")]:
    try:
        df_v  = load_cluster_panel(K=50, lam=lam_v, data_dir=GRID_DIR)
        cr_v  = pivot_returns(df_v, dropna=True)
        cr_vr, _ = rank_clusters(cr_v)
        idx   = cr_vr.index.intersection(cr_ranked.index)
        hl_v  = cr_vr.loc[idx].iloc[:,-1] - cr_vr.loc[idx].iloc[:,0]
        rows3.append(hl_row(label_v,
                            cr_vr.loc[idx].iloc[:,0],
                            cr_vr.loc[idx].iloc[:,-1],
                            hl_v))
    except FileNotFoundError:
        print(f"  {label_v} panel not found — skipping")

# Factor-based benchmarks
for fname, col in [("MKT","Market"),("SMB","Size"),("HML","Value"),
                   ("CMA","Investment"),("MOM","Momentum"),("LIQ","Liquidity")]:
    if fname in fd_c.columns:
        f = fd_c[fname]
        rows3.append(hl_row(col, f.clip(upper=0), f.clip(lower=0), f))

table3 = pd.DataFrame(rows3).set_index("")
print("\n=== Table 3 ===")
print(table3.to_string())
save_table(table3, "Table3_HL_Comparison")
print("Tables 2 and 3 saved.")
```

# Notebook 10 — Figures 15 & 16: Mean-Variance Efficient Frontiers

**Data needed:**
- `data/grid/cluster_month_panel_K_50_lambda_1000000.csv`
- `data/factor_returns.csv`
- `data/mebm_25.csv`          — Fama-French 25 ME-BM portfolios
- `data/benchmark_sorts.csv`  — Uni-Sort 150, Bi-Sort 285, Ind49

**`mebm_25.csv`** format: `date, P11, P12, ..., P55` (25 columns, Ken French website)  
**`benchmark_sorts.csv`** format: `date, US_1, US_2, ..., BS_1, ..., IND_1, ...`

**Output:**
- Figure 15: Scatter plots in (mean, std) and (CAPM alpha, beta) space for two subsamples
- Figure 16: MVE frontier comparison full sample 1977-2020

```
import os, sys, warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from scipy.linalg import solve
warnings.filterwarnings("ignore")

REPO_ROOT = os.path.abspath(os.path.join(os.getcwd(), ".."))
sys.path.insert(0, REPO_ROOT)
from utils.data_utils import (load_cluster_panel, pivot_returns, rank_clusters,
                               load_factor_data, save_figure, GRID_DIR, DATA_DIR)
from utils.portfolio_utils import mve_frontier

plt.rcParams.update({"font.family":"serif","font.size":9,
                      "axes.spines.top":False,"axes.spines.right":False})
RIDGE = 1e-3
```

```
# ── Load baseline cluster returns ─────────────────────────────────
df = load_cluster_panel(K=50, lam=1_000_000, data_dir=GRID_DIR)
cr = pivot_returns(df, dropna=True)
cr_ranked, _ = rank_clusters(cr)
print(f"Cluster returns: {cr_ranked.shape}")

# Load factors
try:
    factors = load_factor_data()
    print(f"Factors: {factors.shape} cols: {list(factors.columns)[:6]}...")
except FileNotFoundError:
    print("factor_returns.csv not found"); factors = pd.DataFrame(index=cr_ranked.index)

# Load ME-BM 25
try:
    mebm = pd.read_csv(os.path.join(DATA_DIR,"mebm_25.csv"), parse_dates=["date"]).set_index("date")
    print(f"ME-BM 25: {mebm.shape}")
except FileNotFoundError:
    print("mebm_25.csv not found — using proxy"); mebm = None

# Load benchmark sorts
try:
    bench = pd.read_csv(os.path.join(DATA_DIR,"benchmark_sorts.csv"),
                        parse_dates=["date"]).set_index("date")
    print(f"Benchmark sorts: {bench.shape}")
except FileNotFoundError:
    print("benchmark_sorts.csv not found"); bench = None

# Common dates
common = cr_ranked.index.intersection(factors.index) if len(factors.columns) > 0 else cr_ranked.index
cr_c   = cr_ranked.loc[common]
fd_c   = factors.loc[common] if len(factors.columns) > 0 else pd.DataFrame(index=common)
```

```
# ── Figure 15: Subsample scatter plots ────────────────────────────
sub_ranges = {
    "1977-1998": (cr_c["1977":"1998"], "1977–1998"),
    "1999-2020": (cr_c["1999":"2020"], "1999–2020"),
}
mkt = fd_c["MKT"] if "MKT" in fd_c.columns else pd.Series(0, index=common)

def capm_alpha_beta(ret, mkt):
    idx = ret.index.intersection(mkt.index)
    if len(idx) < 24: return np.nan, np.nan
    y = ret.loc[idx].values
    X = sm.add_constant(mkt.loc[idx].values)
    res = sm.OLS(y, X).fit()
    return res.params[0]*100, res.params[1]

fig15, axes = plt.subplots(2, 2, figsize=(13, 10))

for col_idx, (key, (cr_sub, sub_label)) in enumerate(sub_ranges.items()):
    # Panel A: mean vs std
    ax = axes[0, col_idx]
    c_mean = cr_sub.mean() * 100
    c_std  = cr_sub.std()  * 100
    ax.scatter(c_std, c_mean, color="black", marker="o", s=35, label="Cluster (Ours)", zorder=3)
    
    if mebm is not None:
        mb_sub = mebm.loc[mebm.index.intersection(cr_sub.index)]
        if len(mb_sub) > 12:
            ax.scatter(mb_sub.std()*100, mb_sub.mean()*100,
                       color="#E91E63", marker="^", s=45, label="ME-BM 5×5", zorder=3)
    
    ax.set_xlabel("Std Dev (% / month)"); ax.set_ylabel("Mean Return (% / month)")
    ax.set_title(f"Panel A: Mean vs Std [{sub_label}]", fontweight="bold")
    ax.legend(fontsize=8)
    
    # Panel B: CAPM alpha vs beta
    ax = axes[1, col_idx]
    mkt_sub = mkt.loc[mkt.index.intersection(cr_sub.index)]
    
    c_alpha, c_beta = zip(*[capm_alpha_beta(cr_sub[col], mkt_sub) for col in cr_sub.columns])
    ax.scatter(list(c_beta), list(c_alpha), color="black", marker="o", s=35,
               label="Cluster (Ours)", zorder=3)
    
    if mebm is not None and len(mb_sub) > 12:
        m_alpha, m_beta = zip(*[capm_alpha_beta(mb_sub[c], mkt_sub) for c in mb_sub.columns])
        ax.scatter(list(m_beta), list(m_alpha), color="#E91E63", marker="^", s=45,
                   label="ME-BM 5×5", zorder=3)
    
    ax.axhline(0, color="grey", linewidth=0.7, linestyle="--")
    ax.set_xlabel("CAPM Beta"); ax.set_ylabel("CAPM Alpha (% / month)")
    ax.set_title(f"Panel B: Alpha vs Beta [{sub_label}]", fontweight="bold")
    ax.legend(fontsize=8)

fig15.suptitle("Figure 15: Diversified Cluster Test Assets Across Subsamples",
               fontsize=12, fontweight="bold")
plt.tight_layout()
save_figure(fig15, "Figure15_Subsamples_Scatter")
plt.show()
print("Figure 15 saved.")
```

```
# ── Figure 16: Full-sample MVE frontiers ──────────────────────────
FACTOR_SETS = {
    "CAPM":  (["MKT"],                          "black",  "-.",  1.4),
    "FF3":   (["MKT","SMB","HML"],              "black",  "-",   1.4),
    "FF5":   (["MKT","SMB","HML","RMW","CMA"],  "black",  "-.",  1.4),
    "FFC":   (["MKT","SMB","HML","MOM"],        "black",  ":",   1.4),
    "Q5":    (["MKT","SMB","IA","ROE_q","EG"],  "#555",   "--",  1.4),
    "DHS":   (["MKT","PEAD","FIN"],             "#555",   ":",   1.4),
}

fig16, ax = plt.subplots(figsize=(11, 6.5))

# Cluster frontier (red solid)
stds_c, means_c, sr_c = mve_frontier(cr_c, ridge=RIDGE)
ax.plot(stds_c*100, means_c*100, color="#D32F2F", linewidth=2.8,
        linestyle="-", label=f"Cluster (SR={sr_c:.2f})", zorder=6)
tan_c = np.argmax(means_c/(stds_c+1e-10))
ax.scatter(stds_c[tan_c]*100, means_c[tan_c]*100, color="#D32F2F", s=90, zorder=7)

# Factor model frontiers
for model, (cols, color, ls, lw) in FACTOR_SETS.items():
    avail = [c for c in cols if c in fd_c.columns]
    if len(avail) < len(cols):
        print(f"  Skipping {model}: missing {set(cols)-set(avail)}")
        continue
    frets = fd_c[avail].dropna()
    stds_f, means_f, sr_f = mve_frontier(frets, ridge=RIDGE)
    ax.plot(stds_f*100, means_f*100, color=color, linestyle=ls, linewidth=lw,
            label=f"{model} (SR={sr_f:.2f})")
    tan_f = np.argmax(means_f/(stds_f+1e-10))
    ax.scatter(stds_f[tan_f]*100, means_f[tan_f]*100, color=color, s=50, zorder=5)

# ME-BM 25 benchmark
if mebm is not None:
    mb_c = mebm.loc[mebm.index.intersection(cr_c.index)].dropna(axis=1)
    if mb_c.shape[1] >= 5:
        stds_mb, means_mb, sr_mb = mve_frontier(mb_c, ridge=RIDGE)
        ax.plot(stds_mb*100, means_mb*100, color="#9C27B0", linestyle="--",
                linewidth=1.4, label=f"ME-BM 5×5 (SR={sr_mb:.2f})")

# Benchmark sorts (Uni-Sort, Bi-Sort, Ind49) 
if bench is not None:
    for label, prefix, color, ls in [
        ("Uni-Sort 150","US_","#FF9800","--"),
        ("Bi-Sort 285", "BS_","#FF9800",":"),
        ("Ind49",       "IND_","#4CAF50","--")
    ]:
        cols_ = [c for c in bench.columns if c.startswith(prefix)]
        if len(cols_) > 2:
            b_c = bench[cols_].loc[bench.index.intersection(cr_c.index)].dropna(axis=1)
            stds_b, means_b, sr_b = mve_frontier(b_c, ridge=RIDGE)
            ax.plot(stds_b*100, means_b*100, color=color, linestyle=ls,
                    linewidth=1.4, label=f"{label} (SR={sr_b:.2f})")

# Latent factors (PCA5, RPPCA5)
for lname, pref, color in [("PCA5","PC","#5C6BC0"),("RPPCA5","RPC","#80DEEA")]:
    cols_ = [f"{pref}{i}" for i in range(1,6)]
    avail = [c for c in cols_ if c in fd_c.columns]
    if len(avail) >= 3:
        stds_l, means_l, sr_l = mve_frontier(fd_c[avail].dropna(), ridge=RIDGE)
        ax.plot(stds_l*100, means_l*100, color=color, linestyle=":",
                linewidth=1.4, label=f"{lname} (SR={sr_l:.2f})")

ax.set_xlabel("Portfolio Std Dev (% / month)")
ax.set_ylabel("Mean Excess Return (% / month)")
ax.set_title("Figure 16: Mean-Variance Efficient Frontiers — Full Sample (1977–2020)",
             fontweight="bold")
ax.legend(fontsize=7.5, loc="upper left", ncol=2, framealpha=0.92)
ax.set_xlim(left=0)
plt.tight_layout()
save_figure(fig16, "Figure16_MVE_Frontier")
plt.show()
print("Figure 16 saved.")
```

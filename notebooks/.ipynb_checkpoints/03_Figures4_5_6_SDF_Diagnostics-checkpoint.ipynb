# Notebook 03 — Figures 4, 5, 6: SDF Diagnostics

**Data needed:** `data/grid/cluster_month_panel_K_50_lambda_1000000.csv`  
**Output:**
- `Figure4_Cumulative_SDF_Return.{pdf,png}`
- `Figure5_SDF_Return_vs_Loading.{pdf,png}`  (corr = -0.916)
- `Figure6_Rolling_SDF_Sharpe.{pdf,png}`

```
import os, sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

REPO_ROOT = os.path.abspath(os.path.join(os.getcwd(), ".."))
sys.path.insert(0, REPO_ROOT)
from utils.data_utils import (load_cluster_panel, pivot_returns, rank_clusters,
                               ann_sharpe, save_figure, GRID_DIR)
from utils.portfolio_utils import estimate_hj_sdf, rolling_hj_sdf

plt.rcParams.update({"font.family":"serif","font.size":10,
                      "axes.spines.top":False,"axes.spines.right":False})

SDF_WINDOW = 72
RIDGE      = 1e-3

# NBER recession shading
RECESSIONS = [("1980-01","1980-07"),("1981-07","1982-11"),
              ("1990-07","1991-03"),("2001-03","2001-11"),
              ("2007-12","2009-06")]
```

```
# ── Load and prepare ─────────────────────────────────────────────
df = load_cluster_panel(K=50, lam=1_000_000, data_dir=GRID_DIR)
cr = pivot_returns(df, dropna=True)
cr_ranked, _ = rank_clusters(cr)

print(f"Returns matrix: {cr_ranked.shape} | {cr_ranked.index[0].date()} → {cr_ranked.index[-1].date()}")

# Full-sample SDF weights
b_full, _ = estimate_hj_sdf(cr_ranked, ridge=RIDGE)
mean_rets  = cr_ranked.mean()

# Rolling OOS SDF returns
print(f"Computing rolling OOS SDF (window={SDF_WINDOW}m)...")
sdf_oos = rolling_hj_sdf(cr_ranked, window=SDF_WINDOW, ridge=RIDGE)
print(f"OOS SDF: {len(sdf_oos)} obs | {sdf_oos.index[0].date()} → {sdf_oos.index[-1].date()}")
```

```
# ── Figure 4: Cumulative SDF return ──────────────────────────────
fig, ax = plt.subplots(figsize=(10, 4))
cum_sdf = (1 + sdf_oos).cumprod() - 1   # in %: cumulative from 0

ax.plot(sdf_oos.index, cum_sdf * 100, color="black", linewidth=1.5)
ax.axhline(0, color="grey", linewidth=0.7, linestyle="--")

for s, e in RECESSIONS:
    ax.axvspan(pd.Timestamp(s), pd.Timestamp(e),
               color="grey", alpha=0.15, linewidth=0)

ax.set_ylabel("OOS Cumulative Return (%)")
ax.set_xlabel("")
ax.set_title("Figure 4: Out-of-Sample Cumulative SDF Return", fontweight="bold")
ax.xaxis.set_major_locator(mdates.YearLocator(4))
ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y"))
plt.tight_layout()
save_figure(fig, "Figure4_Cumulative_SDF_Return")
plt.show()
```

```
# ── Figure 5: Mean return vs SDF loading scatter ─────────────────
sdf_loadings = pd.Series(b_full, index=cr_ranked.columns)
corr = np.corrcoef(mean_rets.values, sdf_loadings.values)[0, 1]
print(f"Correlation mean return vs SDF loading: {corr:.3f}  (paper: -0.916)")

fig, ax = plt.subplots(figsize=(6.5, 5.5))
sc = ax.scatter(mean_rets.values * 100, sdf_loadings.values,
                c=np.arange(len(mean_rets)), cmap="RdYlGn",
                s=55, edgecolors="grey", linewidths=0.4, zorder=3)

# Label each cluster
for i, (label, x, y) in enumerate(zip(cr_ranked.columns,
                                        mean_rets.values*100,
                                        sdf_loadings.values)):
    num = int(label[1:])   # L01 → 1
    ax.annotate(f"${num}$", (x, y), fontsize=6.5,
                ha="center", va="center", color="black")

# Trend line
z = np.polyfit(mean_rets.values*100, sdf_loadings.values, 1)
xr = np.linspace(mean_rets.min()*100, mean_rets.max()*100, 200)
ax.plot(xr, np.poly1d(z)(xr), "k--", linewidth=1, alpha=0.5)
ax.text(0.05, 0.93, f"corr = {corr:.3f}", transform=ax.transAxes,
        fontsize=10, color="darkred")

plt.colorbar(sc, ax=ax, label="Cluster rank (1=lowest, 50=highest)")
ax.set_xlabel("Mean Return (% per month)")
ax.set_ylabel("SDF Loading $b_k$")
ax.set_title("Figure 5: Portfolio Returns and SDF Loadings", fontweight="bold")
plt.tight_layout()
save_figure(fig, "Figure5_SDF_Return_vs_Loading")
plt.show()
```

```
# ── Figure 6: Rolling Sharpe ratio with 95% CI ───────────────────
# Rolling Sharpe
roll_mean = sdf_oos.rolling(SDF_WINDOW).mean()
roll_std  = sdf_oos.rolling(SDF_WINDOW).std()
roll_sr   = (roll_mean / roll_std * np.sqrt(12)).dropna()

# Bootstrap 95% CI on rolling SR
def boot_sr_se(block, n_boot=500):
    boot = [np.random.choice(block, len(block), replace=True) for _ in range(n_boot)]
    srs  = [b.mean()/b.std()*np.sqrt(12) for b in boot]
    return np.std(srs)

np.random.seed(42)
se_vals = {}
for end_date in roll_sr.index[::3]:   # every 3rd point for speed
    loc = sdf_oos.index.get_loc(end_date)
    block = sdf_oos.iloc[max(0, loc-SDF_WINDOW):loc].values
    if len(block) >= SDF_WINDOW:
        se_vals[end_date] = boot_sr_se(block, n_boot=300)
se_series = pd.Series(se_vals).reindex(roll_sr.index).interpolate()

fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(roll_sr.index, roll_sr.values, color="black", linewidth=1.8,
        label=f"Rolling Sharpe ({SDF_WINDOW}m)")
ax.fill_between(roll_sr.index,
                roll_sr - 1.96 * se_series,
                roll_sr + 1.96 * se_series,
                color="grey", alpha=0.25, label="95% CI")
ax.axhline(0, color="grey", linewidth=0.7, linestyle="--")
for s, e in RECESSIONS:
    ax.axvspan(pd.Timestamp(s), pd.Timestamp(e),
               color="grey", alpha=0.12, linewidth=0)
ax.set_ylabel("Annualised Sharpe Ratio")
ax.set_title("Figure 6: Rolling SDF Sharpe Ratio with 95% Confidence Band",
             fontweight="bold")
ax.xaxis.set_major_locator(mdates.YearLocator(4))
ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y"))
ax.legend(fontsize=9)
print(f"SR range: [{roll_sr.min():.2f}, {roll_sr.max():.2f}]  mean: {roll_sr.mean():.2f}")
plt.tight_layout()
save_figure(fig, "Figure6_Rolling_SDF_Sharpe")
plt.show()
print("All three figures saved.")
```

# Notebook 08 — Figures 7 & 8: Industry Decomposition of Dynamic Clusters

**Data needed:** `data/firm_panel.csv` with `SIC` column

**Output:**
- Figure 7: Leading SIC industry per cluster over time (with Manufacturing)
- Figure 8: Leading SIC industry per cluster over time (Manufacturing excluded)

```
import os, sys, warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
warnings.filterwarnings("ignore")

REPO_ROOT = os.path.abspath(os.path.join(os.getcwd(), ".."))
sys.path.insert(0, REPO_ROOT)
from utils.data_utils import (load_cluster_panel, pivot_returns, rank_clusters,
                               load_firm_panel, save_figure, GRID_DIR)

plt.rcParams.update({"font.family":"serif","font.size":8})
```

```
# ── SIC → industry group mapping ─────────────────────────────────
def sic_to_industry(sic):
    if pd.isna(sic): return "Unknown"
    s = int(sic)
    if   s < 1000:  return "Agriculture, Forestry"
    elif s < 1500:  return "Mining"
    elif s < 1800:  return "Construction"
    elif s < 4000:  return "Manufacturing"
    elif s < 5000:  return "Transportation, Communications"
    elif s < 5200:  return "Wholesale Trade"
    elif s < 6000:  return "Retail Trade"
    elif s < 7000:  return "Finance, Insurance"
    elif s < 8000:  return "Services"
    elif s < 9000:  return "Unknown"
    else:           return "Nonclassifiable"

INDUSTRY_COLORS = {
    "Retail Trade":                 "#2196F3",
    "Manufacturing":                "#9E9E9E",
    "Finance, Insurance":           "#FF9800",
    "Transportation, Communications":"#4CAF50",
    "Mining":                       "#795548",
    "Unknown":                      "#FFEB3B",
    "Services":                     "#F44336",
    "Agriculture, Forestry":        "#8BC34A",
    "Wholesale Trade":              "#9C27B0",
    "Construction":                 "#00BCD4",
    "Nonclassifiable":              "#FF69B4",
}
```

```
# ── Load data ─────────────────────────────────────────────────────
df_cl = load_cluster_panel(K=50, lam=1_000_000, data_dir=GRID_DIR)
cr    = pivot_returns(df_cl, dropna=True)
cr_r, rank_map = rank_clusters(cr)

try:
    firm = load_firm_panel()
    if "SIC" not in firm.columns:
        raise ValueError("firm_panel.csv missing 'SIC' column")
    firm["cluster_ranked"] = firm["cluster"].map(rank_map)
    firm["industry"]       = firm["SIC"].apply(sic_to_industry)
    print(f"Firm panel: {firm.shape}")
except (FileNotFoundError, ValueError) as e:
    print(f"Cannot produce Figures 7 & 8: {e}")
    firm = None
```

```
# ── Build leading-industry matrix (cluster × year) ────────────────
def build_leading_industry_matrix(firm_df, exclude=None):
    """Returns (cluster × year) matrix of leading industry string."""
    df = firm_df.copy()
    if exclude:
        df = df[~df["industry"].isin(exclude)]
    
    df["year"] = pd.to_datetime(df["year_month"]).dt.year
    rows = {}
    for (cl, yr), grp in df.groupby(["cluster_ranked","year"]):
        counts = grp["industry"].value_counts()
        rows[(cl, yr)] = counts.index[0] if len(counts) > 0 else "Unknown"
    
    s = pd.Series(rows)
    s.index = pd.MultiIndex.from_tuples(s.index, names=["cluster","year"])
    return s.unstack("year")

def plot_industry_heatmap(mat, title, fname, colors=INDUSTRY_COLORS):
    clusters = [f"L{i:02d}" for i in range(1, 51)]
    mat = mat.reindex(clusters)
    years = sorted(mat.columns)
    
    all_inds = sorted(set(mat.values.flatten()) - {np.nan, None, "nan"})
    ind_idx  = {ind: i for i, ind in enumerate(all_inds)}
    
    # Numeric matrix for imshow
    num_mat = mat.applymap(lambda x: ind_idx.get(x, -1) if pd.notna(x) else -1)
    
    fig, ax = plt.subplots(figsize=(20, 10))
    cmap_colors = [colors.get(ind, "#CCCCCC") for ind in all_inds]
    from matplotlib.colors import ListedColormap, BoundaryNorm
    cmap_list = ListedColormap(["white"] + cmap_colors)
    bounds    = list(range(-1, len(all_inds) + 1))
    norm      = BoundaryNorm(bounds, cmap_list.N)
    
    ax.imshow(num_mat.values, aspect="auto", cmap=cmap_list, norm=norm,
              interpolation="nearest")
    
    ax.set_yticks(range(len(clusters))); ax.set_yticklabels(clusters, fontsize=6)
    year_ticks = [i for i,y in enumerate(years) if y % 5 == 0]
    ax.set_xticks(year_ticks)
    ax.set_xticklabels([years[i] for i in year_ticks], fontsize=8)
    ax.set_xlabel("Year"); ax.set_ylabel("Cluster")
    ax.set_title(title, fontsize=11, fontweight="bold")
    
    patches = [mpatches.Patch(color=colors.get(ind,"#CCC"), label=ind)
               for ind in all_inds]
    ax.legend(handles=patches, loc="lower center", ncol=4, fontsize=7,
              bbox_to_anchor=(0.5, -0.18), framealpha=0.9, title="Industry")
    
    plt.tight_layout()
    save_figure(fig, fname)
    return fig

if firm is not None:
    print("Building industry matrices (may take a minute)...")
    mat_all  = build_leading_industry_matrix(firm)
    mat_nomf = build_leading_industry_matrix(firm, exclude=["Manufacturing"])
    
    fig7 = plot_industry_heatmap(mat_all,  "Figure 7: Industry Decomposition of Dynamic Clusters",
                                  "Figure7_Industry_Heatmap_All")
    plt.show()
    
    fig8 = plot_industry_heatmap(mat_nomf, "Figure 8: Industry Decomposition (Manufacturing Excluded)",
                                  "Figure8_Industry_Heatmap_NoMfg")
    plt.show()
    print("Figures 7 and 8 saved.")
```

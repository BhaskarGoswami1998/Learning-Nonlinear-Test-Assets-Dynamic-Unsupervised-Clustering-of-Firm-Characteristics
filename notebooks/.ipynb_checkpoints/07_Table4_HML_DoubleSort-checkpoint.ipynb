# Notebook 07 — Table 4: HML Double-Sort Returns

**Data needed:** `data/firm_panel.csv`  
Columns required: `year_month, permno, cluster, ret, me, [45 chars]`

**Output:** `Table4_HML_DoubleSort.{csv,tex}`

For each of 45 characteristics × each cluster, this notebook:
1. Within each (date, cluster), splits firms at the median of the characteristic
2. Computes VW return spread (High − Low)
3. Reports means with NW t-statistics for L1-L5 and L46-L50
4. Counts statistically significant positive/negative clusters (#PosSig, #NegSig, PropSig)

```
import os, sys, warnings
import numpy as np
import pandas as pd
warnings.filterwarnings("ignore")

REPO_ROOT = os.path.abspath(os.path.join(os.getcwd(), ".."))
sys.path.insert(0, REPO_ROOT)
from utils.data_utils import (load_cluster_panel, pivot_returns, rank_clusters,
                               load_firm_panel, nw_tstat, stars, save_table,
                               GRID_DIR, DATA_DIR, CHARS_45)
NW_LAGS = 3
```

```
# ── Load data ─────────────────────────────────────────────────────
df_cl = load_cluster_panel(K=50, lam=1_000_000, data_dir=GRID_DIR)
cr    = pivot_returns(df_cl, dropna=True)
cr_r, rank_map = rank_clusters(cr)

try:
    firm = load_firm_panel()
    firm["cluster_ranked"] = firm["cluster"].map(rank_map)
    char_cols = [c for c in CHARS_45 if c in firm.columns]
    print(f"Firm panel: {firm.shape}, {len(char_cols)} chars available")
    print(f"Date range: {firm['year_month'].min()} → {firm['year_month'].max()}")
except FileNotFoundError:
    print("firm_panel.csv not found — cannot produce Table 4.")
    print("Provide data/firm_panel.csv with cols: year_month, permno, cluster, ret, me, [chars]")
    firm = None; char_cols = []
```

```
# ── Compute HML returns per char × cluster ────────────────────────
TARGET_CLUSTERS = [f"L{i:02d}" for i in range(1,6)] + [f"L{i:02d}" for i in range(46,51)]
ALL_CLUSTERS    = list(cr_r.columns)

if firm is not None:
    def vw_hml_series(sub_df, char):
        """Monthly HML series within a cluster subset."""
        monthly = []
        for dt, grp in sub_df.groupby("year_month"):
            grp = grp.dropna(subset=[char,"ret","me"])
            if len(grp) < 4: continue
            med = grp[char].median()
            hi  = grp[grp[char] >  med]
            lo  = grp[grp[char] <= med]
            if len(hi) == 0 or len(lo) == 0: continue
            r_hi = (hi["ret"] * hi["me"]).sum() / hi["me"].sum()
            r_lo = (lo["ret"] * lo["me"]).sum() / lo["me"].sum()
            monthly.append(r_hi - r_lo)
        return pd.Series(monthly)
    
    print("Computing HML returns (this takes a while)...")
    
    # Results: char → {cluster_label → "coef***"}
    hml_results = {}
    sig_counts   = {}   # char → {posig, negsig}
    
    for char in char_cols:
        print(f"  {char}...", end="", flush=True)
        cell_results = {}
        pos_sig = neg_sig = 0
        
        for cl in ALL_CLUSTERS:
            sub = firm[firm["cluster_ranked"] == cl]
            if len(sub) < 50: cell_results[cl] = np.nan; continue
            hml = vw_hml_series(sub, char)
            if len(hml) < 12: cell_results[cl] = np.nan; continue
            m, se, t = nw_tstat(hml, lags=NW_LAGS)
            cell_results[cl] = m
            if not pd.isna(t):
                if t >=  1.96: pos_sig += 1
                if t <= -1.96: neg_sig += 1
        
        # Format target clusters with stars
        disp = {}
        for cl in TARGET_CLUSTERS:
            if cl not in cell_results or pd.isna(cell_results[cl]):
                disp[cl] = ""
                continue
            sub = firm[firm["cluster_ranked"]==cl]
            hml = vw_hml_series(sub, char)
            m, se, t = nw_tstat(hml, lags=NW_LAGS)
            disp[cl] = f"{m:.3f}{stars(t)}"
        
        hml_results[char] = disp
        sig_counts[char]  = {"#PosSig": pos_sig, "#NegSig": neg_sig,
                              "PropSig": round((pos_sig+neg_sig)/50, 2)}
        print(" done")
    
    hml_df  = pd.DataFrame(hml_results).T[TARGET_CLUSTERS]
    sig_df  = pd.DataFrame(sig_counts).T
    table4  = pd.concat([hml_df, sig_df], axis=1)
    table4.index.name = "Feature"
    print("\n=== Table 4 (first 5 rows) ===")
    print(table4.head())
    save_table(table4, "Table4_HML_DoubleSort")
    print("Table 4 saved.")
```
